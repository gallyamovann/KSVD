```python
# версия с битами
import math

import h5py
import numpy as np
import csv
from PIL import Image

def convert_weather_value_to_px(left_border, value):
    if isinstance(value, float):
        value *= 100
        left_border *= 100
        left_border, value = int(left_border), int(value)
    b = 0 if value >= 0 else 255
    shifted_value = abs(value) - left_border
    r = shifted_value // 256
    g = shifted_value % 256
    return [0, 0, r, g]


def convert_snp_to_px(placed, value, bits):
    r, g, b = 0, 0, 0
    if value == 1:
        r = (1 << placed) | bits
        g = 0
    elif value == 0:
        r = 0
        g = 0
    elif value == 2:
        r = (1 << placed) | bits
        g = (1 << placed) | bits
    return [math.sqrt(r) if math.sqrt(r) <= 255 else 255, math.sqrt(g) if math.sqrt(g) <= 255 else 255], bits


# def get_weather_information(filename):
#     with h5py.File(filename, "r") as f:
#         weather_info = []
#         weather_groups = list(f.keys())
#         weather_groups[0], weather_groups[3] = weather_groups[3], weather_groups[0]
#         # ['month', 'doy', 'geo_id', 'dl', 'rain', 'srad', 'tmax', 'tmin', 'year']
#         for group in weather_groups:
#             tmp = []
#             for dset in f[group]:
#                 tmp.append(dset[0])
#             weather_info.append(np.array(tmp))
#         weather_info = np.array(weather_info).transpose()
#         return weather_info


def get_weather_information(filename):
    df = pd.read_csv(filename)
    weather_info = []
    weather_groups = df.columns.values.tolist()
    # ['month', 'doy', 'geo_id', 'dl', 'rain', 'srad', 'tmax', 'tmin', 'year']
    for group in weather_groups:
        tmp = []
        for dset in df[group]:
            tmp.append(dset)
        weather_info.append(np.array(tmp))
    weather_info = np.array(weather_info).transpose()
    return weather_info


def get_chickpea_information(filename):
    df = pd.read_csv(filename)
    for key in ('month', 'gr_names'):
        df = df.drop(key, axis=1)
    plant_groups = df.columns.values.tolist()
    print(plant_groups)
    chickpea_info = []
    # ['doy', 'geo_id', 'gr_covar', 'response', 'species', 'year']
    for group in plant_groups:
        tmp = []
        for dset in df[group]:
            if group == 'species':
                tmp.append(dset.decode('UTF-8'))
            elif group == 'gr_covar':
                if (dset[0] == 1.):
                    dset[0] = 2
                if (dset[3] == 1.):
                    dset[3] = 2
                if (dset[6] == 1.):
                    dset[6] = 2
                if (dset[9] == 1.):
                    dset[9] = 2
                if (dset[12] == 1.):
                    dset[12] = 2
                if (dset[15] == 1.):
                    dset[15] = 2
                if (dset[1] == 1.):
                    dset[1] = 0
                if (dset[4] == 1.):
                    dset[4] = 0
                if (dset[7] == 1.):
                    dset[7] = 0
                if (dset[10] == 1.):
                    dset[10] = 0
                if (dset[13] == 1.):
                    dset[13] = 0
                if (dset[16] == 1.):
                    dset[16] = 0
                if (dset[2] == 1.):
                    dset[2] = 1
                if (dset[5] == 1.):
                    dset[5] = 1
                if (dset[8] == 1.):
                    dset[8] = 1
                if (dset[11] == 1.):
                    dset[11] = 1
                if (dset[14] == 1.):
                    dset[14] = 1
                if (dset[17] == 1.):
                    dset[17] = 1
                tmp.append(dset)
            else:
                tmp.append(dset)
        chickpea_info.append(np.array(tmp))
    chickpea_info = list(map(list, zip(*chickpea_info)))
    plants = []
    for chickpea in chickpea_info:
        plants.append(chickpea)
    return plants


def transform(filename_chickpea, filename_weather, path_to_save, csv_class):
    tmp_weather_first = get_weather_information(filename_weather)
    weather = []
    for i in range(len(tmp_weather_first)):
        weather.append(tmp_weather_first[i])
    all_chickpea = get_chickpea_information(filename_chickpea)
    num_of_image = 1
    repeat = []
    with open(csv_class, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL)
        for plant in all_chickpea:
            pixels1 = []
            pixels2 = []
            snp_pixels = []
            placed = 0
            temp_bit = 0
            for snp in plant[2]:
                temp_array, temp_bit = convert_snp_to_px(placed, snp, temp_bit)
                snp_pixels.append(temp_array)
                placed += 1
            index_snp = 0
            for day in weather:
                if (day[-1] == plant[-1]) and (day[2] == plant[1]) and (plant[0] - 5 <= day[1] < plant[0] + 20):
                    day_pixels = []
                    for i in range(3, 8):
                        tmp = convert_weather_value_to_px(0, day[i])
                        tmp[0] = snp_pixels[index_snp][0]
                        tmp[1] = snp_pixels[index_snp][1]
                        day_pixels.append(tmp)
                    if index_snp > 16:
                        index_snp = 0
                    index_snp += 1
                    pixels1.append(day_pixels)
                    pixels2.append(day_pixels)
            pixels = np.zeros((25, 25, 4))
            for i in range(25):
                for j in range(25):
                    pixels[i][j] = pixels1[i][j % 5]
            pixels = np.array(pixels, dtype=np.uint8)
            new_image = Image.fromarray(pixels, mode='RGBA').rotate(90, expand=True)
            new_image.save(path_to_save + "image" + str(num_of_image) + ".png")
            print("image" + str(num_of_image) + ".png")
            writer.writerow(["image" + str(num_of_image) + ".png", str(plant[3])])  # only response
            num_of_image += 1


path_to_save = "images_test_b/"
chickpea = "chickpea-ssm-snp.csv"
weather = "weather.csv"
csv_response = "response_test_b.csv"
transform(chickpea, weather, path_to_save, csv_response)
```
